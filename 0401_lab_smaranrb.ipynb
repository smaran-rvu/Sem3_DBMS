2. Print the min, max, and mean heights of boys and girls for the whole class and also by section.  


```python
pd.read_sql("SELECT SEX, MAX(Height_cm) MAX_HEIGHT, MIN(Height_cm) as MIN_HEIGHT, AVG(Height_cm) AVG_HEIGHT FROM STUDENTS GROUP BY SEX", con)
```

```python
pd.read_sql("SELECT Section, MAX(Height_cm) MAX_HEIGHT, MIN(Height_cm) as MIN_HEIGHT, AVG(Height_cm) AVG_HEIGHT FROM STUDENTS GROUP BY Section", con)
```
