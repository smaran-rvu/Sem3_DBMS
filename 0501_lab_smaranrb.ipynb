```python
sql = """
  UPDATE Students_alt
     SET Home_State = Master_state.State_code
     FROM Master_state
        WHERE Master_state.State_name = Students_alt.Home_State
"""
con.execute(sql)
sqlcon = """
    ALTER TABLE Students_alt
    RENAME COLUMN Home_State TO State_id;
"""
con.execute(sqlcon)
con.commit()
```

```python
sql = """
  UPDATE Students_alt
     SET Home_Town = Master_city.District_code
     FROM Master_city
        WHERE Master_city.City_name = Students_alt.Home_town
"""
con.execute(sql)
sqlcon = """
    ALTER TABLE Students_alt
    RENAME COLUMN Home_Town TO Town_id;
"""
con.execute(sqlcon)
con.commit()
```

```python
sql = """
  UPDATE Students_alt
     SET Mother_Tongue = Master_language.Code
     FROM Master_language
        WHERE Master_language.Language = Students_alt.Mother_Tongue
"""
con.execute(sql)
sqlcon = """
    ALTER TABLE Students_alt
    RENAME COLUMN Mother_tongue TO Language_id;
"""
con.execute(sqlcon)
con.commit()
```

```python
sql = """
  UPDATE Students_alt
     SET Elective = Master_elective.Elective_ID
     FROM Master_elective
        WHERE Master_elective.Name = Students_alt.Elective
"""
con.execute(sql)
sqlcon = """
    ALTER TABLE Students_alt
    RENAME COLUMN Elective TO Elective_id;
"""
con.execute(sqlcon)
con.commit()
```
